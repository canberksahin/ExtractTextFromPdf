```python
# Gerekli kütüphaneleri yükle
!pip install pdfplumber
import pdfplumber
import re
from google.colab import files

# Kullanıcıdan bir PDF dosyası yüklemesini iste
uploaded = files.upload()
pdf_filename = list(uploaded.keys())[0]

def temizle_metin(text):
    """
    PDF'den çıkarılan metni temizleyen fonksiyon:
    - Sayfa numaralarını akıllıca kaldırır
    - Özel karakterleri düzeltir
    - İçindekiler kısmını filtreler
    - Fazla boşlukları kaldırır
    """
    if not text:
        return ""

    # "Page X", "Sayfa X" veya sadece "X" olarak geçen sayfa numaralarını temizle
    text = re.sub(r'\b(Page|Sayfa)\s*\d+\b', '', text)  # Page 12, Sayfa 15 gibi
    text = re.sub(r'^\d+$', '', text, flags=re.MULTILINE)  # Sadece "11" gibi tek başına satırda duranları kaldır

    # İçindekiler kısmını tespit edip filtreleme
    if "İÇİNDEKİLER" in text.upper():
        return ""  # Eğer sayfa içindekiler sayfasıysa metni almıyoruz

    # Unicode karakter düzeltmeleri
    text = text.replace("\xa0", " ")
    text = text.replace("ﬁ", "fi").replace("ﬂ", "fl")

    # Fazla boşlukları düzelt
    text = re.sub(r'\s+', ' ', text).strip()

    return text

clean_text = ""
prev_text = ""  # Önceki sayfanın sonu ile bu sayfanın başını birleştirmek için

with pdfplumber.open(pdf_filename) as pdf:
    for page in pdf.pages:
        if page.extract_text():
            current_text = temizle_metin(page.extract_text())

            # Eğer önceki sayfa boş değilse, metinleri boş satır olmadan birleştir
            if prev_text:
                clean_text += " " + current_text  # Yeni satır eklemiyoruz
            else:
                clean_text += current_text

            prev_text = current_text  # Bir sonraki sayfa için önceki metni kaydet

output_filename = "clean_extracted_text.txt"
with open(output_filename, "w", encoding="utf-8") as f:
    f.write(clean_text)

files.download(output_filename)

print(clean_text[:1000])
```
